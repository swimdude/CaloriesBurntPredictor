```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
from sklearn import metrics
```

**Reading Files**

```python
calories = pd.read_csv('calories.csv')
exercise_data = pd.read_csv('exercise.csv')
```

```python
calories.head()
```

```python
exercise_data.head()
```

```python
# axis=1 makes it so the calories data is loaded adter the last column
calories_data = pd.concat([exercise_data, calories['Calories']], axis = 1)
calories_data.head()
```

```python
#shows rows and columns
calories_data.shape
```

```python
calories_data.info()
```

```python
#checking for missing values
calories_data.isnull().sum()
```

```python
#Converting text data to numerical values

calories_data.replace({"Gender":{'male':0, 'female':1}}, inplace =True) #inplace makes the change throughout
```

```python
calories_data.head()
```

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
X
```

```python
Y
```

**Splitting Data**

```python
# test_size is set to 20%    
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# train has 80% of data and test has 20% of data
X.shape, X_train.shape, X_test.shape
```

**Model Training & Evaluation**

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

```python
# Prediction on Test Data

test_data_prediction = model.predict(X_test)
```

```python
test_data_prediction
```

```python
# mean absolute error

mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error = ", mae)
```
